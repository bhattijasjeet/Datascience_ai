```python
import os
os.environ["HDF5_USE_FILE_LOCKING"] = "FALSE"

# Import image module
from tensorflow.keras.preprocessing import image

# Import ResNet50 as myModel
from tensorflow.keras.applications.resnet50 import ResNet50 as myModel

# Import preprocess_input and decode_predictions
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions

# Load the model with ImageNet weights
model = myModel(weights="imagenet")
```

```python
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np

# Image path
img_path = '/cxldata/dlcourse/lion.jpg'

# Load the image with target size 224 x 224
img = image.load_img(img_path, target_size=(224, 224))

# Display the image
plt.imshow(np.asarray(img))
plt.axis('off')  # optional: hide axis ticks
plt.show()
```

```python
# Convert the PIL image to numpy array
x = image.img_to_array(img)

# Model expects a batch, so wrap it in an array
x = np.array([x])

# Preprocess the input
x = preprocess_input(x)

# Predict the class probabilities
preds = model.predict(x)

# Decode and print top 3 predictions
print('Predicted:', decode_predictions(preds, top=3)[0])
```
